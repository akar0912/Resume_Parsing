```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\AlishaRavikaKar\Desktop\res\resume\resumes_filter15.csv")
```

```python
df['Experience2'] = df['Experience'].str.extract('(\d+)').astype(float)

df['Exp_weight'] = df['Experience2'].apply(lambda x: 1 if x < 5 else(2.5 if x<10 and x >= 5 else(3.5 if x > 10 else 0.0)))
```

```python
df
```

```python
import numpy as np
df['Degree'] = df['Degree'].str.strip("['']")


degree_wt = {'B.Tech':1,'Bachelors In Engineering':1,'Bachelor Of Technology':1,\
'Btech':1,'M.Sc':2,'M.Tech':3,'Masters':3,'':0}


df['Degree_Weight'] = [degree_wt[i] if i in degree_wt.keys() else np.nan for i in df["Degree"]]
df['ranking_score'] = (df['python_count']*3)+(df['vba_count']*2)+(df['excel_count']*1.5)\
         +(df['sql_count']*2)+(df['Degree_Weight'])+(df['Exp_weight'])
df['rank'] = df['ranking_score'].rank(method = "dense", ascending = False)
df = df.sort_values(['ranking_score'],ascending =False).reset_index(drop=True)

```
